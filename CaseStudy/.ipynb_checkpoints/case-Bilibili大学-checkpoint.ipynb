## 数据清理
由于数据中右很多空格和回车，而且有些播放量带‘万’，有些则以数字形式存在于表格中，因此我们首先对数据进行清理，使观看次数、弹幕数量为数字形式。

```python
## 去掉空格
data['描述'] = data['描述'].map(str.strip)
data['观看次数'] = data['观看次数'].map(str.strip)
data['弹幕数'] = data['弹幕数'].map(str.strip)
data['上传时间'] = data['上传时间'].map(str.strip)
```

```python
## 编写函数，将xx万转换成数字
def containwan(string):
    flag = string.find('万')
    if flag != -1:
        string = string[:flag]
        strint = float(string)*10000
        return strint
    else:
        return float(string)
data['观看次数'] = data['观看次数'].map(containwan)
data['弹幕数'] = data['弹幕数'].map(containwan)
```

```python
## 把上传时间更改为时间格式
from datetime import datetime
data['上传时间'] = data['上传时间'].map(lambda x:datetime.strptime(x,'%Y-%m-%d'))
## 把年和月单独提取出来
data['年'] = data['上传时间'].dt.year
data['月'] = data['上传时间'].dt.month
```

## 视频收藏数和硬币数、弹幕数、观看次数之间的联系

```python
import sys
import math
sys.setrecursionlimit(5000)
```

```python
def nplog(x):
    if x <= 0:
        return 0
    else:
        return math.log(x)
```

```python
data['对数观看'] = np.log(data['观看次数'])
data['对数弹幕'] = data['弹幕数'].apply(nplog)
data['对数点赞'] = data['like'].apply(nplog)
data['对数评论'] = data['reply'].apply(nplog)
data['对数分享'] = data['share'].apply(nplog)
data['对数硬币'] = data['coin'].apply(nplog)
```

```python
g = sns.pairplot(data[["对数收藏", "对数观看",'对数评论','对数分享','对数硬币','对数弹幕','对数点赞']],kind = 'reg' )
```

&emsp;&emsp;根据图像，可以大致看出课程的几个变量间的联系和分布。其中，收藏数和其他几个反应用户活跃度的指标之间的线性关系都较弱，数据的方差比较大，相对而言收藏数和硬币数、点赞数的关联比较强。观看数和分享数、弹幕数的线性相关最强；评论数和弹幕数线性相关性最强；分享数和观看数的线性相关最强；硬币数和点赞数线性相关最强；弹幕数和观看、分享数线性相关比较强，和点赞数几乎没有线性相关；点赞数和硬币数线性相关最强。<br/>
&emsp;&emsp;因此，可以大致将这些反应用户活跃度的信息分成两类，一类是分享数、观看数、弹幕数、评论数；一类是点赞数、硬币数、收藏数。<br/>
&emsp;&emsp;为了进一步分析互动指标之间的关联，可以使用因子分析方法，观察这些变量之间的关联。

```python
original = data[["对数收藏", "对数观看",'对数评论','对数分享','对数硬币','对数弹幕','对数点赞']]
##标准化
preprocess = (original - original.mean())/original.std()
##相关系数矩阵
corr = preprocess.corr()
##计算特征值和特征向量
(evalue,evector)= np.linalg.eig(corr)
eig = pd.DataFrame()
eig['特征'] = original.columns
eig['特征值'] = evalue
eig.head()
```

```python
## 确定公共因子个数,解释度到80%即可
for k in range(7):
    if eig['特征值'][:k].sum()/eig['特征值'].sum()>=0.8:
        print(k)
        break
```

```python
## 构造因子载荷矩阵
col1=list(math.sqrt(evalue[0])*evector[:,0]) 
col2=list(math.sqrt(evalue[1])*evector[:,1]) 
A = pd.DataFrame([col1,col2]).T
A.columns = ['因子1','因子2']
A
```

```python
h = np.zeros(7)
D = np.mat(np.eye(7))
A = np.mat(A)
for i in range(7):
    #A的元的行平方和
    a=A[i,:]*A[i,:].T 
    #计算变量X共同度,描述全部公共因子F对变量X_i的总方差所做的贡献，及变量X_i方差中能够被全体因子解释的部分
    h[i]=a[0,0]  
    #因为自变量矩阵已经标准化后的方差为1，即Var(X_i)=第i个共同度h_i + 第i个特殊因子方差
    D[i,i]=1-a[0,0] 
```

```python
def varimax(Phi, gamma = 1.0, q =20, tol = 1e-6): #定义方差最大旋转函数
    p,k = Phi.shape #给出矩阵Phi的总行数，总列数
    R = np.eye(k) #给定一个k*k的单位矩阵
    d=0
    for i in range(q):
        d_old = d
        Lambda = np.dot(Phi, R)
        u,s,vh = np.linalg.svd(np.dot(Phi.T,np.asarray(Lambda)**3 - (gamma/p) * np.dot(Lambda, np.diag(np.diag(np.dot(Lambda.T,Lambda)))))) 
        R = np.dot(u,vh)#构造正交矩阵R
        d = np.sum(s)#奇异值求和
        if d_old!=0 and d/d_old < 1+tol:
            break
    return np.dot(Phi, R)#返回旋转矩阵Phi*R

rotation_mat=varimax(A)#调用方差最大旋转函数
rotation_mat=pd.DataFrame(rotation_mat)#数据框化
rotation_mat.columns = ['公共因子1','公共因子2']

rotation_mat
```

&emsp;&emsp;根据旋转过后的因子载荷矩阵，我们可以看到，第一公共因子中载荷主要是'对数收藏','对数硬币','对数点赞'，他们表示了观众本身对于这个视频的认可程度其中最大值为‘对数点赞’，载荷为-0.88;第二公共因子中，主要起作用的变量是'对数观看','对数分享','对数弹幕'，他们表示了观众对这个视频的互动程度其中最大载荷为‘对数弹幕’,载荷为0.93。而‘对数评论’指标在两个公共因子中的载荷差异不显著，主要偏向于互动。这也体现了在B站的多个体现视频热度的指标中，最有效的指标分别为‘点赞数’和‘弹幕数’。而目前在B站官网上，可供视频排序的指标仅有‘弹幕数’、‘收藏数’和‘观看数’，因此建议增加一个按照‘点赞数’排序的接口，更方便让观众找到有价值、有意义的课程。

## 文本信息挖掘
&emsp;&emsp;为了进一步研究课程的主题，我们想提取高频词并绘制词云图

```python
import jieba.analyse
segments = []
stopwords = [line.strip() for line in open('中文停用词.txt', 'r', encoding='utf-8').readlines()] 
stopwords += ['课程','教程','公开课','公开','开课','视频','教学','【','】','老师','课时','小时',
              '分钟','零基','算机','学会','视频教程','学院']
for index, row in data.iterrows():
    content = row[2]
   # words = jieba.analyse.textrank(content, topK=50,withWeight=False)
    words = jieba.cut(content,cut_all = True)
    for word in words:
        # 记录全局分词
        word = word.strip()
        if word not in stopwords:
            segments.append({'word':word,'count':1})
```

```python
dfwords = pd.DataFrame(segments)
words = dfwords.groupby('word').sum()
words['word'] = words.index
words.columns = ['count','word']
words = dict(filter(lambda x:len(x[0]) >= 2 ,words.iterrows()))
words = pd.DataFrame(words).T
words[(words['count']>15) ].sort_values(ascending = False,by = 'count')
```

## 武林争霸：高校 vs 野生视频
为了细致地比较高校公开课视频和其他野生课程之间的差异，我们决定利用标题挖掘的信息，新建一些变量并对这些变量进行k-means聚类，得到一些典型的课程类型。(注:*此处课程分类参考了Coursera在线教育平台上的学科划分)

```python
data['大学'] = data['题目'].map(lambda x : find_course(x,['大学','耶鲁','清华','北大','麻省理工',
                                                      '大學','北师大','斯坦佛','伯克利','交大','中科大']))
data['入门'] = data['题目'].map(lambda x : find_course(x,['基础','入门','菜鸟','速成','初级','零']))

data['考'] = data['题目'].map(lambda x : find_course(x,['考']))
```

```python
def divide_class(x):
    info = ['计算机','数据','操作系统','深度学习','机器学习','算法','Python','python','编程','程序','C语言','Java','智能'
            ,'编译','机械','cad','Hadoop','Web','linux']
    base = ['语文','汉语','数学','化学','生物','物理','力学','电','代数','解剖','医','生理','药理','儿科','概率']
    popu = ['心理','经济','社会','教育','刑法','政府','金灿荣','民法','博弈','逻辑','新闻']
    art = ['艺术','编曲','音乐','史','文学','哲','Pr','视频','pr','PS','Adobe','乐理','钢琴'
           ,'红楼梦','美学','审美','影片','诗','写作','主义','设计','PPT','word','AE','唱歌','五线谱']
    busi = ['会计','商','金融','市场','领导','营销','管理','演讲','思维','画','水彩']
    lang = ['日本','韩国','法国','西班牙','德国','美国','英国','语']
    tup = str(x)
    if find_course(x,popu):
        return 4
    elif find_course(x,info):
        return 2
    elif find_course(x,base):
        return 1
    elif find_course(x,art):
        return 6
    elif find_course(x,busi):
        return 5
    elif find_course(x,lang):
        return 3
    return 7
data['题材'] = data[['题目','描述']].apply(lambda x :divide_class(x),axis = 1)
```

```python
data.head()
#data[['题目','描述']]
```

```python
college = data[(data['大学']==1) & (data['考']==0)] #筛选大学公开课
other = data[(data['大学']==0) & (data['copyright']==1) &(data['考']==0)] #筛选不是大学的原创视频
(len(college),len(other)) #两者数据量差不多，具有可比性
```

```python
from sklearn.cluster import KMeans

kcollege = KMeans(n_clusters=5,random_state=0).fit(college[['题材','入门','对数点赞','对数弹幕']])
kother = KMeans(n_clusters=4,random_state=0).fit(other[['题材','入门','对数点赞','对数弹幕']])
tcollege = kcollege.cluster_centers_
tother = kother.cluster_centers_
(tcollege,tother)
```

&emsp;&emsp;根据聚类的结果，可以将高校公开课分成5类，偏向人文类的课程(题材=6)，课程入门级的较少，弹幕数和点赞数都较低；偏向商务类(题材=5)课程，课程相对较难，但是弹幕数和点赞数都较高；对于偏向语言类(题材=3)的课程，课程难度偏容易，课程点赞数最低，弹幕数中等偏低；对于偏向基础学科类（题材=1）和信息技术类（题材=2）的课程，如果课程难度低，弹幕数和点赞数都很高；如果课程难度系数最高，则点赞数偏低，弹幕数最低，可能是因为内容过于专业，受众比较小，而且大家都真正忙于学习，无心互动 <br/>
&emsp;&emsp;可以将野生课程分成4类，偏向人文艺术类的课程(题材=6)，如果难度系数低，弹幕数和点赞数都较高，远高于同类高校公开课程；如果非入门级的课程多，则弹幕数和点赞数都很低，与高校公开课的情况类似；对于偏向语言类(题材=3)的课程，入门级的课程比例与同类高校公开课类似，课程弹幕数和点赞数较低；对于偏向信息技术类（题材=2）的课程，课程难度较容易，弹幕数和点赞数都相对最高，和高校公开课类似；同时，野生视频区相对缺少基础学科类视频，可能是因为专业性太强，不方便自制。同时，我们可以看到，野生区的视频整体上难度比高校公开课简单很多。<br/>
&emsp;&emsp;因此，我们可以总结，对于想爆红的UP主来说，可以自制一些简单易懂或者容易上手的信息技术类视频或者人文艺术类视频或者搬运一些高校的商务类视频和简单易懂的基础学科类视频。
